# Modelo Kmeans (Twitter)

## Prueba #1 con caracteristicas: 'op', 'co' y 'ex'

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Cargar el conjunto de datos
df = pd.read_csv('/content/drive/MyDrive/usuarios-twitter.csv')

# Caractersticas para el modelo
features = df[['op', 'co', 'ex']]

# Normalizar los datos
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Mtodo del codo para encontrar el nmero ptimo de clusters
inertias = []
silhouettes = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_scaled)
    inertias.append(kmeans.inertia_)
    silhouettes.append(silhouette_score(features_scaled, kmeans.labels_))

# Graficar la suma de las distancias cuadradas
plt.figure(figsize=(8, 4))
plt.plot(range(2, 11), inertias, marker='o')
plt.title('Mtodo del codo')
plt.xlabel('Nmero de clusters')
plt.ylabel('Inercia')
plt.show()

# Graficar los scores de silhouette
plt.figure(figsize=(8, 4))
plt.plot(range(2, 11), silhouettes, marker='o')
plt.title('Silhouette Scores')
plt.xlabel('Nmero de clusters')
plt.ylabel('Silhouette Score')
plt.show()

# Elegir el nmero ptimo de clusters
optimal_clusters = 4
kmeans_final = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans_final.fit(features_scaled)

# Asignar las etiquetas de los clusters a los datos originales
df['Cluster'] = kmeans_final.labels_

```

# Modelo RNA

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

# Simulacin de datos
X, y = make_blobs(
    n_samples=500,
    n_features=2,
    centers=3,
    cluster_std=1.2,
    shuffle=True,
    random_state=0
)

# Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Pipeline para estandarizar los datos y luego aplicar MLPClassifier
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('mlp', MLPClassifier(max_iter=1000, random_state=0))
])

# Parmetros para GridSearchCV
parameter_space = {
    'mlp__hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 100)],
    'mlp__activation': ['tanh', 'relu'],
    'mlp__solver': ['sgd', 'adam'],
    'mlp__alpha': [0.0001, 0.05],
    'mlp__learning_rate': ['constant','adaptive'],
}

# K-Fold cross-validator
kfold = StratifiedKFold(n_splits=5)

# GridSearchCV
clf = GridSearchCV(pipeline, parameter_space, n_jobs=-1, cv=kfold)
clf.fit(X_train, y_train)

# Mejor modelo encontrado
print('Mejores parmetros encontrados:\n', clf.best_params_)

# Precisin del mejor modelo
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Precisin del mejor modelo:', accuracy)
```

## Conclusin

Segn los resultados obtenidos de la ejecucin, la configuracin de la RNA que gener el mejor accuracy en la tarea de clasificacin de las muestras simuladas en tres grupos es la siguiente:

* Funcin de activacin: relu (unidad lineal rectificada)
* Tamao de capas ocultas: (50,) (una sola capa oculta con 50 neuronas)
* Tasa de aprendizaje: constante
* Algoritmo de optimizacin (solver): sgd (descenso de gradiente estocstico)

La precisin del mejor modelo es del 87%. Esto significa que el modelo fue capaz de clasificar correctamente el 87% de las muestras en el conjunto de datos de prueba.
